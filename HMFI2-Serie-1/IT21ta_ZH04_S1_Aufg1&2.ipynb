```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook 
from mpl_toolkits.mplot3d import Axes3D

def W(v0, alpha):
    return (v0**2 * np.sin(2*alpha))/9.81

[x,y] = np.meshgrid(np.linspace(0, 100), np.linspace(0, np.pi/2))
z = W(x, y)

fig = plt.figure(1)
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y ,z)

plt.title('Gitter')
ax.set_xlabel('Abwurfgeschwindigkeit')
ax.set_ylabel('Abwurfwinkel')
ax.set_zlabel('Wurfdistanz')

plt.show()

#Die maximale Wurfdistanz wird bei einem Abwurfwinkel von 45Â° erreicht
```

```python
fig = plt.figure(2)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z)

plt.title('FlÃ¤che')
ax.set_xlabel('Abwurfgeschwindigkeit')
ax.set_ylabel('Abwurfwinkel')
ax.set_zlabel('Wurfdistanz')

plt.show()
```

```python
from matplotlib import cm

fig = plt.figure(3)
cont = plt.contour(x, y, z, cmap=cm.coolwarm)
fig.colorbar(cont, shrink=0.5, aspect=5)

plt.title('HÃ¶henlinien')
plt.xlabel('Abwurfgeschwindigkeit')
plt.ylabel('Abwurfwinkel')

plt.show()
```

Teilaufgabe b.)

$p = $ Druck in $\frac{N}{m^2}$

$V = $ Volumen in $m^3$

$T = $ absolute Temperatur in Kelvin

$R = $ Gaskonstante in $\frac{J}{molK}$

```python
R = 8.31

def function_p(V, T):
    return (R * T) / V 

def function_V(p, T):
    return (R*T) / p

def function_T(p, V): 
    return (p * V) / R

[V1,T1] = np.meshgrid(np.linspace(0, 0.2), np.linspace(0, np.e**4))
p_result = function_p(V1, T1)

[p1, T2] = np.meshgrid(np.linspace(np.e**4, np.e**5), np.linspace(0, np.e**4))
V_result = function_V(p1, T2)

[p2, V2] = np.meshgrid(np.linspace(np.e**4, np.e**6), np.linspace(0, 10))
T_result = function_T(p2, V2)

fig = plt.figure(4)
fig.set_size_inches(15, 8)
ax = fig.add_subplot(221, projection='3d')
ax.plot_wireframe(V1, T1 ,p_result)

ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_wireframe(p1, T2, V_result)

ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_wireframe(p2, V2, T_result)

plt.title('Gitter')
ax.set_xlabel('Volumen in $m^3$')
ax.set_ylabel('Absolute Temperatur in Kelvin')
ax.set_zlabel('Druck in $N / m^2$')

ax2.set_xlabel('Druck in $N / m^2$')
ax2.set_ylabel('Absolute Temperatur in Kelvin')
ax2.set_zlabel('Volumen in $m^3$')

ax3.set_xlabel('Druck in $N / m^2$')
ax3.set_ylabel('Volumen in $m^3$')
ax3.set_zlabel('Absolute Temperatur in Kelvin')

plt.show()
```

```python
fig = plt.figure(5)
fig.set_size_inches(15, 8)

ax = fig.add_subplot(221, projection='3d')
ax.plot_surface(V1, T1 ,p_result)

ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(p1, T2, V_result)

ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_surface(p2, V2, T_result)

plt.title('FlÃ¤che')
ax.set_xlabel('Volumen in $m^3$')
ax.set_ylabel('Absolute Temperatur in Kelvin')
ax.set_zlabel('Druck in $N / m^2$')

ax2.set_xlabel('Druck in $N / m^2$')
ax2.set_ylabel('Absolute Temperatur in Kelvin')
ax2.set_zlabel('Volumen in $m^3$')

ax3.set_xlabel('Druck in $N / m^2$')
ax3.set_ylabel('Volumen in $m^3$')
ax3.set_zlabel('Absolute Temperatur in Kelvin')

plt.show()
```

```python
def w(x, t):
    return np.sin(x+t)

def v(x, t):
    return np.sin(x+t) + np.cos(2*x+2*t)

[x,t] = np.meshgrid(np.linspace(0, 8), np.linspace(0, 8))
w_result = w(x, t)

[x2,t2] = np.meshgrid(np.linspace(0, 8), np.linspace(0, 8))
v_result = v(x2, t2)

fig = plt.figure(8)
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, t, w_result)

ax.set_xlabel('x')
ax.set_ylabel('t')

plt.show()
```

```python
fig = plt.figure(9)
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x2, t2, v_result)

plt.title('Gitter')
ax.set_xlabel('x')
ax.set_ylabel('t')
```
