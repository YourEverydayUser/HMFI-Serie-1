Aufgabe 2.a)

```python
import sympy as sp

x, y = sp.symbols(' x y ')
f1 = x**2 / 186**2 - y**2/(300**2-186**2) - 1
f2 = (y-500)**2/279**2 - (x-300)**2/(500**2 - 279**2) - 1
p1 = sp.plot_implicit(sp.Eq(f1, 0), (x, -2000 ,2000), (y, -2000, 2000))
p2 = sp.plot_implicit(sp.Eq(f2, 0), (x, -2000 ,2000), (y, -2000, 2000))
p1.append(p2[0])
p1.show()
```

Die vier Schnittpunkte sind ungefähr (-1300, 1600), (-200, 100), (800, 900), (300, 200)

Aufgabe 2.b)

```python
import numpy as np

f = sp.Matrix([f1, f2])
Df = f.jacobian([x, y])
f = sp.lambdify([([x], [y])], f, 'numpy')
Df = sp.lambdify([([x], [y])], Df, 'numpy')
```

```python
startVectors = []
startVectors.append([[-1300.], [1600.]])
startVectors.append([[-200.], [100.]])
startVectors.append([[-800.], [900.]])
startVectors.append([[-300.], [200.]])

epsilon = 10**-5
for x in startVectors:
    while np.linalg.norm(f(x), 2) > epsilon:
        d = np.linalg.solve(Df(x), -f(x))
        x += d
    print(x)
```
