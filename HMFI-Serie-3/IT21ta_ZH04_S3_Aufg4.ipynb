```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mat

def h(x):
    return np.sqrt(100*x**2 - 200 * x + 99)

def konditionszahl(x):
    return 100 * (x**2 -x)

x_values2 = np.arange(0.75, 1.3, 10**-3)
x_values = np.arange(1.1+3*10**-16, 1.1+500*10**-16, 3*10**-16)
x_values3 = np.arange(0.7, 1.3, 10**-4)
y_values = []

for x in x_values:
    y_values.append(h(x))

k = []

for x in x_values2:
    k.append(konditionszahl(x))

y2 = []

for x in x_values3:
    y2.append(h(x))

    
fig, axs = plt.subplots(2, 2)
fig.set_size_inches(20, 5)

axs[0, 0].set_xlim(1.10, 1.1000000000001)
axs[0, 0].plot(x_values, y_values)

axs[0, 1].set_xlim(1.1, 1.3)
axs[0, 1].plot(x_values2, k)

axs[1, 0].set_xlim(0.8, 1.3)
axs[1, 0].plot(x_values3, y2)




```
