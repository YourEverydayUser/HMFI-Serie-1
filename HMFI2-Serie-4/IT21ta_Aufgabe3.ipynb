```python
import numpy as np
import matplotlib.pyplot as plt

x = [1981, 1984, 1989, 1993, 1997, 2000, 2001, 2003, 2004, 2010]
y = [0.5, 8.2, 15, 22.9, 36.6, 51, 56.3, 61.8, 65, 76.7]

order = len(x)-1

polynomial = np.polyfit(x, y, order)
x_values = np.arange(1975, 2020, 0.1)
y_values = np.polyval(polynomial, x_values)

plt.xlim([1981, 2010])
plt.ylim([0.5, 76.7])
plt.plot(x, y, marker='o', color='b')
plt.plot(x_values, y_values)
```

Das berechnete Polynom geht nicht exakt durch alle Datenpunkte.

```python
x2 = x-np.mean(x)

polynomial2 = np.polyfit(x2, y, order)
y_values = np.polyval(polynomial2, x_values)

plt.xlim([1981, 2010])
plt.ylim([0.5, 76.7])
plt.plot(x, y, marker='o', color='b')
plt.plot(x_values, y_values)
```
