```python
from cProfile import label
import numpy as np
import matplotlib.pyplot as plt

def create_function(a, xmin, xmax):
    x = np.arange(xmin, xmax, 0.1)
    if len(a) == 0:
        raise Exception("At least one coefficient must be given.")
    y = 0
    temp_exp = len(a) - 1
    for i in range(len(a)):
        y = y + a[i] * x**temp_exp
        temp_exp = temp_exp -1
    return y

def create_derivative(a, xmin, xmax):
    x = np.arange(xmin, xmax, 0.1)
    if len(a) == 0:
        raise Exception("At least one coefficient must be given.")
    y = 0
    temp_exp = len(a) - 1
    for i in range(len(a)):
        y = y + temp_exp * a[i] * x**(temp_exp - 1)
        temp_exp = temp_exp -1
    return y

def create_integral(a, xmin, xmax):
    x = np.arange(xmin, xmax, 0.1)
    if len(a) == 0:
        raise Exception("At least one coefficient must be given.")
    y = 0
    temp_exp = len(a) - 1
    for i in range(len(a)):
        y = y + (1/(temp_exp+1)) * a[i] * x**(temp_exp + 1)
        temp_exp = temp_exp -1
    return y

def plot_function(a, xmin, xmax):
    x = np.arange(xmin, xmax, 0.1)
    y = create_function(a, xmin, xmax)
    der = create_derivative(a, xmin, xmax)
    int = create_integral(a, xmin, xmax)
    plt.plot(x,y, label = "f(x)")
    plt.plot(x,der, label = "f'(x)")
    plt.plot(x,int, label = "F(x)")
    plt.ylim(-10,10)
    plt.legend()
    return(x, y, der, int)
    


plot_function([2,1], -10, 10)
```
