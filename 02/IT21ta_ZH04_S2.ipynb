```python
import numpy as np
import matplotlib.pyplot as plt
from regex import F
import sympy

def f1(x):
    return x**7 - 14*x**6 + 84*x**5 - 280*x**4 + 560*x**3 - 672*x**2 + 448*x - 128
    
def f2(x):
    return (x-2)**7

def IT21ta_ZH04_S2_Aufg2_a(xmin, xmax):

    x_axis = np.arange(xmin, xmax, 0.000039999)

    f1_values = []
    f2_values = []
  

    for x in x_axis:
        f1_values.append(f1(x))
        f2_values.append(f2(x))

    

    plt.figure(figsize=(10, 10))
    plt.plot(x_axis, f1_values, label = "f1(x)")
    plt.plot(x_axis, f2_values, label = "f2(x)")
    plt.xlabel("x-Values")
    plt.ylabel("y-Values")
    
    plt.legend()
    plt.grid()

    
    return(x_axis, f1_values, f2_values)

IT21ta_ZH04_S2_Aufg2_a(1.99, 2.01)

```

```python
def f3(x):
    return x / (np.sin(1 + x) - np.sin(1))

def IT21ta_ZH04_S2_Aufg2_b(xmin, xmax):

    x_axis = np.arange(xmin, xmax, 10**(-17))

    f3_values = []

    for x in x_axis:
        f3_values.append(f3(x))
    

    plt.figure(figsize=(10, 10))
    plt.plot(x_axis, f3_values)
    plt.xlabel("x-Values")
    plt.ylabel("y-Values")
    
    plt.legend()
    plt.grid()
        
    return(x_axis, f3_values)

IT21ta_ZH04_S2_Aufg2_b(-10**-14, 10**-14)
```
