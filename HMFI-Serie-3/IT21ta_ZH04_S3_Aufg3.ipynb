```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mat

def f(x):
    return 5 / np.cbrt(2*x**2)

def g(x):
    return 10**5 * (2*np.e)**-x/100

def h(x):
    return (5**4*x) * (2**-6*x)

def h2(x):
    return ((10**2*x)/(2**5*x))**2

def iterate(function):
    xvalues = []
    yvalues = []

    for x in range(100):
        xvalues.append(x)
        yvalues.append(function(x))

    return xvalues, yvalues

fig, axs = plt.subplots(3, 2)
fig.set_size_inches(18.5, 10.5)

#f(x)
axs[0, 0].plot(iterate(f)[0], iterate(f)[1])
axs[0, 1].loglog(iterate(f)[0], iterate(f)[1])

#g(x)
axs[1, 0].plot(iterate(g)[0], iterate(g)[1])
axs[1, 1].semilogy(iterate(g)[0], iterate(g)[1])

#h(x)
axs[2, 0].plot(iterate(h)[0], iterate(h)[1])
axs[2, 1].loglog(iterate(h)[0], iterate(h)[1])




```
