```python
from audioop import ratecv
import timeit
import numpy as np
import matplotlib.pyplot as plt

def fact_rec(n):
# y = fact_rec(n) berechnet die Fakultät von n als fact_rec(n) = n * fact_rec(n -1) mit fact_rec(0) = 1
# Fehler, falls n < 0 oder nicht ganzzahlig
    import numpy as np
    if n < 0 or np.trunc(n) != n:
        raise Exception('The factorial is defined only for positive integers')
    if n <=1:
        return 1
    else:
        return n*fact_rec(n-1)

def fact_for(n):
    result = 1
    for i in range(1,n+1):
        result = result * i
    return result

t1= timeit.repeat("fact_rec(500)", "from __main__ import fact_rec", number=100)
t2= timeit.repeat("fact_for(500)", "from __main__ import fact_for", number=100)
print(t1)
print(t2)

# 1) Der for-Loop scheint um einen Faktor 10 schneller zu sein (0.01 vs 0.1). Das liegt daran, dass bei Rekursion immer wieder die momentane Position auf den Stack geschrieben
# werden muss, wenn erneut in die Funktion gegangen wird. Dies sind zusätzliche Rechenschritte.
# 2) Wahrscheinlich gibt es eine Grenze, da z.B. Integer nur limitierte Stellen/Bits für die Darstellung zur Verfügung haben. Andere Überlegung: Speicher ist immer limitiert. 
# D.h. bei einer unendlich grossen Zahl wird der Speicher eines Computers sein Limit erreicht haben. Ausserdem ist die Fakultätsfunktion nur für positive Integer definiert. D.h.
# Zahlen mit Kommastellen können nicht als Funktionswerte mitgegeben werden.

for i in range(190, 200):
    print(fact_rec(i))
    print(fact_for(i))
    print("")
    
```
