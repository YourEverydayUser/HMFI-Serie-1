```python
from cProfile import label
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.01)
y = x**5 - 5 * x**4 - 30 * x**3 + 110 * x**2 + 29 * x - 105

derv_y = 5 * x**4 - 20 * x**3 - 90 * x**2 + 220 * x + 29
integral_y = 1/6 * x**6 - x**5 - 30/4 * x**4 + 110/3 * x**3 + 29/2 * x**2 + x

plt.plot(x, y)
plt.plot(x, derv_y)
plt.plot(x, integral_y)
plt.xlim(-7.5, 7.5)
plt.ylim(-1000, 1000)
plt.legend(["f(x)", "f'(x)", "F(x)"])
plt.grid()
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("f(x) = x^5 − 5x^4 − 30x^3 + 110x^2 + 29x − 105")

plt.show
```
