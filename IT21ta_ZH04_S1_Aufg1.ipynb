```python
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.arange(-10, 10, 0.01)
y_values = []
dy_values = []
Y_values = []

def baseFunction(x):
    returnValue = x**5 - 5 * x**4 - 30 * x**3 + 110 * x**2 - 29 * x - 105
    return returnValue

def firstDerivative(x):
    returnValue = 5 * x**4 - 20 * x**3 - 90 * x**2 + 220 * x + 29
    return returnValue

def firstAntiderivative(x):
    returnValue = Y = 1/6 * x**6 - x**5 - 30/4 * x**4 + 110/3 * x**3 + 29/2 * x**2 + x
    return returnValue

for x in x_axis:
    y = baseFunction(x)
    y_values.append(y)
    
for x in x_axis:
    dy = firstDerivative(x)
    dy_values.append(dy)
    
for x in x_axis:
    Y = firstAntiderivative(x)
    Y_values.append(Y)
    

#plotting    
plt.plot(x_axis, y_values, label = "f(x) = f(x) = x^5 − 5x^4 − 30x^3 + 110x^2 + 29x − 105")
plt.plot(x_axis, dy_values, label = "Derivative of f(x)")
plt.plot(x_axis, Y_values, label = "Antiderivative of f(x)")
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.title("Graphs")
plt.legend()
plt.xlim()
plt.ylim(-2000, 2000)
plt.grid()
plt.show()
```
